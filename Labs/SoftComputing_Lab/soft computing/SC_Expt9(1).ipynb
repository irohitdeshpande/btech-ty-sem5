```python
import numpy as np
import matplotlib.pyplot as plt

def triangular_membership(x, a, b, c):
    return np.maximum(np.minimum((x-a)/(b-a), (c-x)/(c-b)), 0)

def max_membership(x, membership_values):
    return x[np.argmax(membership_values)]

def centroid(x, membership_values):
    return np.sum(x * membership_values) / np.sum(membership_values)

def center_of_sums(x, membership_values):
    return np.sum(x * np.sum(membership_values)) / np.sum(np.sum(membership_values))

def mean_of_maximum(x, membership_values):
    max_value = np.max(membership_values)
    max_indices = np.where(membership_values == max_value)
    return np.mean(x[max_indices])

def weighted_average(x, membership_values):
    return np.sum(membership_values * x) / np.sum(membership_values)

def center_of_largest_area(x, membership_values):
    half_max = np.max(membership_values) / 2
    left_idx = np.argmax(membership_values >= half_max)
    right_idx = len(x) - np.argmax(membership_values[::-1] >= half_max)
    return np.mean(x[left_idx:right_idx])

x = np.linspace(0, 10, 100)
membership_values = triangular_membership(x, 2, 5, 8)

z_max_membership = max_membership(x, membership_values)
z_centroid = centroid(x, membership_values)
z_center_of_sums = center_of_sums(x, membership_values)
z_mean_of_maximum = mean_of_maximum(x, membership_values)
z_weighted_average = weighted_average(x, membership_values)
z_center_of_largest_area = center_of_largest_area(x, membership_values)

plt.figure(figsize=(10, 6))
plt.plot(x, membership_values, label='Fuzzy Set', color='blue')
plt.axvline(z_max_membership, color='r', linestyle='--', linewidth=1.5, label='Max Membership')
plt.axvline(z_centroid, color='g', linestyle='--', linewidth=1.5, label='Centroid')
plt.axvline(z_center_of_sums, color='purple', linestyle='--', linewidth=1.5, label='Center of Sums')
plt.axvline(z_mean_of_maximum, color='orange', linestyle='--', linewidth=1.5, label='Mean of Maximum')
plt.axvline(z_weighted_average, color='brown', linestyle='--', linewidth=1.5, label='Weighted Average')
plt.axvline(z_center_of_largest_area, color='pink', linestyle='--', linewidth=1.5, label='Center of Largest Area')

plt.title('Defuzzification Methods Comparison', fontsize=14)
plt.xlabel('x (Crisp Value)', fontsize=12)
plt.ylabel('Membership Degree', fontsize=12)
plt.xlim([0, 10])
plt.ylim([0, 1.1])
plt.legend(loc='upper right', fontsize=10)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Triangular membership function
def triangular_membership(x, a, b, c):
    return np.maximum(np.minimum((x - a) / (b - a), (c - x) / (c - b)), 0)

# Height Method (Max Membership)
def height_method(x, membership_values):
    return x[np.argmax(membership_values)]

# First of Maxima (FoM)
def first_of_maxima(x, membership_values):
    max_value = np.max(membership_values)
    max_indices = np.where(membership_values == max_value)[0]
    return x[max_indices[0]]

# Last of Maxima (LoM)
def last_of_maxima(x, membership_values):
    max_value = np.max(membership_values)
    max_indices = np.where(membership_values == max_value)[0]
    return x[max_indices[-1]]

# Mean of Maxima (MoM)
def mean_of_maxima(x, membership_values):
    max_value = np.max(membership_values)
    max_indices = np.where(membership_values == max_value)
    return np.mean(x[max_indices])

# Generate x values and fuzzy set
x = np.linspace(0, 10, 100)
membership_values = triangular_membership(x, 2, 5, 8)

# Compute defuzzified values using different Maxima methods
z_height_method = height_method(x, membership_values)
z_first_of_maxima = first_of_maxima(x, membership_values)
z_last_of_maxima = last_of_maxima(x, membership_values)
z_mean_of_maxima = mean_of_maxima(x, membership_values)

# Plot the fuzzy set and defuzzification results
plt.figure(figsize=(10, 6))
plt.plot(x, membership_values, label='Fuzzy Set', color='blue')
plt.axvline(z_height_method, color='r', linestyle='--', linewidth=1.5, label='Height Method')
plt.axvline(z_first_of_maxima, color='g', linestyle='--', linewidth=1.5, label='First of Maxima (FoM)')
plt.axvline(z_last_of_maxima, color='purple', linestyle='--', linewidth=1.5, label='Last of Maxima (LoM)')
plt.axvline(z_mean_of_maxima, color='orange', linestyle='--', linewidth=1.5, label='Mean of Maxima (MoM)')

plt.title('Maxima Defuzzification Methods Comparison', fontsize=14)
plt.xlabel('x (Crisp Value)', fontsize=12)
plt.ylabel('Membership Degree', fontsize=12)
plt.xlim([0, 10])
plt.ylim([0, 1.1])
plt.legend(loc='upper right', fontsize=10)
plt.grid(True)
plt.show()
```
