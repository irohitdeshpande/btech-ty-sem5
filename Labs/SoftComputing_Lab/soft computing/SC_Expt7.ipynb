```python
import math

X = [0.05, 0.10]
W = [0.6, -0.1, -0.3, 0.4, 0.4, 0.1]
Hin = [0, 0,]
Hout = [0, 0]
O = [0, 0]
B = [0.3, 0.5, -0.2]
E = 0
t = 1

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def forward_pass(X, W, Hin, Hout, O):
  Hin[0] = W[0] * X[0] + W[1] * X[1] + B[0]
  Hin[1] = W[2] * X[0] + W[3] * X[1] + B[1]
  Hout[0] = sigmoid(Hin[0])
  Hout[1] = sigmoid(Hin[1])

  O[0] = W[4] * Hout[0] + W[5] * Hout[1] + B[2]
  O[1] = sigmoid(O[0])
  E = 0.5 * (t - O[1]) ** 2
  return Hin, Hout, O, E

print("Inputs: ", X)
print("Weights: ", W)
print("Targets: ", t)
print("Biases: ", B)
Hin, Hout, O, E = forward_pass(X, W, Hin, Hout, O)
print("Hidden Values (input):", Hin)
print("Hidden Values (output):", Hout)
print("Outputs (Calculated, Actual):", O)
print("Error:", E)
```
