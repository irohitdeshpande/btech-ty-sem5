```python
import numpy as np

def mc_culloch_pitts(weights, inputs, threshold):
    weighted_sum = np.dot(weights, inputs)
    return 1 if weighted_sum >= threshold else 0

def not_function(x):
    weights = [-1]
    threshold = 0
    return mc_culloch_pitts(weights, [x], threshold)

def nand_function(x1, x2):
    and_weights = [1, 1]
    and_threshold = 2
    and_output = mc_culloch_pitts(and_weights, [x1, x2], and_threshold)
    return not_function(and_output)

def and_not_function(x1, x2):
    and_weights = [1, 1]
    and_threshold = 2
    not_x1 = not_function(x1)
    return mc_culloch_pitts(and_weights, [not_x1, x2], and_threshold)

print("NAND function outputs:")
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(f"nand({x1}, {x2}) = {nand_function(x1, x2)}")

print("\nAND NOT function outputs:")
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(f"and_not({x1}, {x2}) = {and_not_function(x1, x2)}")
```
