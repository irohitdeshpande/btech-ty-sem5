```python
def get_ip():
    ip = input("Enter IP Address/Slash Bits (e.g., 192.168.1.1/24): ").split("/")
    return ip

def get_subnet_mask(slash_bits):
    subnet_mask = []
    # Create the subnet mask in binary form
    for i in range(4):
        if slash_bits >= 8:
            subnet_mask.append(255)
            slash_bits -= 8
        else:
            subnet_mask.append(256 - 2**(8-slash_bits))
            slash_bits = 0
    return subnet_mask

def get_network_id(ip, subnet_mask):
    network_id = []
    # Perform bitwise AND operation for each octet
    for i in range(4):
        network_id.append(str(int(ip[i]) & subnet_mask[i]))
    return network_id

def display(ip, subnet_mask, network_id, subnet):
    print("IP Address: ", ".".join(ip))
    print("Subnet Mask: ", ".".join(map(str, subnet_mask)))
    print("Network ID: ", ".".join(network_id))
    print("Number of Hosts: ", subnet - 2)  # Subtracting 2 for network and broadcast address

if __name__ == "__main__":
    ip_input = get_ip()
    slash_bits = int(ip_input[-1])
    ip = ip_input[0].split(".")
    subnet = 2**(32 - slash_bits)  # Calculating the number of hosts in the subnet
    subnet_mask = get_subnet_mask(slash_bits)
    network_id = get_network_id(ip, subnet_mask)
    display(ip, subnet_mask, network_id, subnet)
```
