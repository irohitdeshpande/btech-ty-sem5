## OS Experiment 3
## Implementing non-preemptive process scheduling algorithms
## FCFS, SJF, Priority

```python
process_id = []
wt = []
priority = []
bt = []
tat = []
at = []
ready_queue = []

total_time = 0
count = 0
n = 0
avg_wt = 0
avg_tat = 0

flag = True
while flag:
  print("1. FCFS\n2. SJF\n3. Priority\n4. Exit")
  choice = int(input("Enter your choice: "))
  if choice == 4:
    flag = False
    break

  elif choice == 1:
    n = int(input("Enter the number of processes: "))

    for i in range(n):
      process_id.append(i+1)
      bt.append(int(input(f"Enter the burst time for process {i+1}: ")))
      wt.append(0)
      tat.append(0)
      total_time += bt[i]

    for i in range(1,n):
      wt[i] = bt[i-1] + wt[i-1]

    for i in range(n):
      tat[i] = bt[i] + wt[i]

    for i in range(n):
      avg_wt += wt[i]
      avg_tat += tat[i]
    avg_wt /= n
    avg_tat /= n
    print("Process\t\tBurst Time\tWaiting Time\tTurnaround Time")
    for i in range(n):
      print(f"{process_id[i]}\t\t{bt[i]}\t\t{wt[i]}\t\t{tat[i]}")
    print(f"Average Waiting Time: {avg_wt}")
    print(f"Average Turnaround Time: {avg_tat}")
    print()

  elif choice == 2:
    n = int(input("Enter the number of processes: "))

    for i in range(n):
      process_id.append(i+1)
      at.append(int(input(f"Enter the arrival time for process {i+1}: ")))
      bt.append(int(input(f"Enter the burst time for process {i+1}: ")))
      wt.append(0)
      tat.append(0)
      total_time += bt[i]

    for i in range(n):
      for j in range(i+1, n):
        if at[i] > at[j]:
          at[i], at[j] = at[j], at[i]
          bt[i], bt[j] = bt[j], bt[i]
          process_id[i], process_id[j] = process_id[j], process_id[i]
        elif at[i] == at[j]:
          if bt[i] > bt[j]:
            at[i], at[j] = at[j], at[i]
            bt[i], bt[j] = bt[j], bt[i]
            process_id[i], process_id[j] = process_id[j], process_id[i]
        else:
          continue

    wt[0] = 0
    for i in range(1,n):
      wt[i] = bt[i-1] + wt[i-1]
      avg_wt += wt[i]

    for i in range(n):
      tat[i] = bt[i] + wt[i]
      avg_tat += tat[i]

    avg_wt /= n
    avg_tat /= n
    print("Process\t\tArrival Time\tBurst Time\tWaiting Time\tTurnaround Time")
    for i in range(n):
      print(f"{process_id[i]}\t\t{at[i]}\t\t{bt[i]}\t\t{wt[i]}\t\t{tat[i]}")
    print(f"Average Waiting Time: {avg_wt}")
    print(f"Average Turnaround Time: {avg_tat}")
    print()

  elif choice == 3:
    n = int(input("Enter the number of processes: "))

    for i in range(n):
      process_id.append(i+1)
      bt.append(int(input(f"Enter the burst time for process {i+1}: ")))
      priority.append(int(input(f"Enter the priority for process {i+1}: ")))
      wt.append(0)
      tat.append(0)
      total_time += bt[i]

    for i in range(n):
      for j in range(i+1, n):
        if priority[i] > priority[j]:
          priority[i], priority[j] = priority[j], priority[i]
          bt[i], bt[j] = bt[j], bt[i]
          process_id[i], process_id[j] = process_id[j], process_id[i]

    wt[0] = 0
    for i in range(1,n):
      wt[i] = bt[i-1] + wt[i-1]
      avg_wt += wt[i]

    for i in range(n):
      tat[i] = bt[i] + wt[i]
      avg_tat += tat[i]

    avg_wt /= n
    avg_tat /= n
    print("Process\t\tBurst Time\tWaiting Time\tTurnaround Time")
    for i in range(n):
      print(f"{process_id[i]}\t\t{bt[i]}\t\t{wt[i]}\t\t{tat[i]}")
    print(f"Average Waiting Time: {avg_wt}")
    print(f"Average Turnaround Time: {avg_tat}")
    print()

  else:
    print("Invalid choice")
    print()
```
